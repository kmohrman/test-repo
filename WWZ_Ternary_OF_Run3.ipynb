```python
of_file = pd.read_csv('mva_scores_OF_Run3.txt', sep=' ').values
```

```python
#Create arrays for the elements we care about

#WWZ scores
p_wwz_s_wwz_bdtR2 = []
p_wwz_s_wwz_bdtR3 = []
p_zh_s_wwz_bdtR2 = []
p_zh_s_wwz_bdtR3 = []
p_bkg_s_wwz_bdtR2 = []
p_bkg_s_wwz_bdtR3 = []

#ZH scores
p_wwz_s_zh_bdtR2 = []
p_wwz_s_zh_bdtR3 = []
p_zh_s_zh_bdtR2 = []
p_zh_s_zh_bdtR3 = []
p_bkg_s_zh_bdtR2 = []
p_bkg_s_zh_bdtR3 = []

#BKG scores
p_wwz_s_bkg_bdtR2 = []
p_wwz_s_bkg_bdtR3 = []
p_zh_s_bkg_bdtR2 = []
p_zh_s_bkg_bdtR3 = []
p_bkg_s_bkg_bdtR2 = []
p_bkg_s_bkg_bdtR3 = []

#Weights
p_wwz_weights = []
p_zh_weights = []
p_bkg_weights = []
```

```python
for x in range(0,112041):
    # WWZ process
    if of_file[x][0] == 0:
        p_wwz_s_wwz_bdtR3.append(of_file[x][1])
        p_wwz_s_zh_bdtR3.append(of_file[x][2])
        p_wwz_s_bkg_bdtR3.append(of_file[x][3])
        p_wwz_s_wwz_bdtR2.append(of_file[x][4])
        p_wwz_s_zh_bdtR2.append(of_file[x][5])
        p_wwz_s_bkg_bdtR2.append(of_file[x][6])
        p_wwz_weights.append(of_file[x][7])
    # ZH process
    if of_file[x][0] == 1:
        p_zh_s_wwz_bdtR3.append(of_file[x][1])
        p_zh_s_zh_bdtR3.append(of_file[x][2])
        p_zh_s_bkg_bdtR3.append(of_file[x][3])
        p_zh_s_wwz_bdtR2.append(of_file[x][4])
        p_zh_s_zh_bdtR2.append(of_file[x][5])
        p_zh_s_bkg_bdtR2.append(of_file[x][6])
        p_zh_weights.append(of_file[x][7])
    # BKG process
    if of_file[x][0] == 2:
        p_bkg_s_wwz_bdtR3.append(of_file[x][1])
        p_bkg_s_zh_bdtR3.append(of_file[x][2])
        p_bkg_s_bkg_bdtR3.append(of_file[x][3])
        p_bkg_s_wwz_bdtR2.append(of_file[x][4])
        p_bkg_s_zh_bdtR2.append(of_file[x][5])
        p_bkg_s_bkg_bdtR2.append(of_file[x][6])
        p_bkg_weights.append(of_file[x][7])
        
# Convert everything to numpy arrays....
arr_p_wwz_s_wwz_bdtR3 = np.array(p_wwz_s_wwz_bdtR3)
arr_p_wwz_s_zh_bdtR3 = np.array(p_wwz_s_zh_bdtR3)
arr_p_wwz_s_bkg_bdtR3 = np.array(p_wwz_s_bkg_bdtR3)
arr_p_wwz_s_wwz_bdtR2 = np.array(p_wwz_s_wwz_bdtR2)
arr_p_wwz_s_zh_bdtR2 = np.array(p_wwz_s_zh_bdtR2)
arr_p_wwz_s_bkg_bdtR2 = np.array(p_wwz_s_bkg_bdtR2)
arr_p_wwz_weights = np.array(p_wwz_weights)

arr_p_zh_s_wwz_bdtR3 = np.array(p_zh_s_wwz_bdtR3)
arr_p_zh_s_zh_bdtR3 = np.array(p_zh_s_zh_bdtR3)
arr_p_zh_s_bkg_bdtR3 = np.array(p_zh_s_bkg_bdtR3)
arr_p_zh_s_wwz_bdtR2 = np.array(p_zh_s_wwz_bdtR2)
arr_p_zh_s_zh_bdtR2 = np.array(p_zh_s_zh_bdtR2)
arr_p_zh_s_bkg_bdtR2 = np.array(p_zh_s_bkg_bdtR2)
arr_p_zh_weights = np.array(p_zh_weights)

arr_p_bkg_s_wwz_bdtR3 = np.array(p_bkg_s_wwz_bdtR3)
arr_p_bkg_s_zh_bdtR3 = np.array(p_bkg_s_zh_bdtR3)
arr_p_bkg_s_bkg_bdtR3 = np.array(p_bkg_s_bkg_bdtR3)
arr_p_bkg_s_wwz_bdtR2 = np.array(p_bkg_s_wwz_bdtR2)
arr_p_bkg_s_zh_bdtR2 = np.array(p_bkg_s_zh_bdtR2)
arr_p_bkg_s_bkg_bdtR2 = np.array(p_bkg_s_bkg_bdtR2)
arr_p_bkg_weights = np.array(p_bkg_weights)

print("Number of WWZ events = ",arr_p_wwz_weights.size)
print("Number of ZH events = ",arr_p_zh_weights.size)
print("Number of BKG events = ",arr_p_bkg_weights.size)

#arr_p_wwz_weights /= 4.
#arr_p_zh_weights /= 4.
#arr_p_bkg_weights /= 4.
```

```python
# Let's use a formula to (roughly) calculate the significance
def calculateFOM(S,B):
        term1 = S+B
        term2 = np.log(1. + (S/B))

        return np.sqrt(2.*((term1*term2)-S))
```

```python
wwz_yields_r2_bdt = [0.,0.,0.,0.]
wwz_yields_r3_bdt = [0.,0.,0.,0.]
zh_yields_r2_bdt = [0.,0.,0.,0.]
zh_yields_r3_bdt = [0.,0.,0.,0.]
bkg_yields_r2_bdt = [0.,0.,0.,0.]
bkg_yields_r3_bdt = [0.,0.,0.,0.]

# WWZ loop
for x in range (0,arr_p_wwz_weights.size):
    #R2 BDT
    if (arr_p_wwz_s_wwz_bdtR2[x] > arr_p_wwz_s_zh_bdtR2[x]):
        if (arr_p_wwz_s_bkg_bdtR2[x] <= 0.13):
            wwz_yields_r2_bdt[0] += arr_p_wwz_weights[x]
        else:
            wwz_yields_r2_bdt[1] += arr_p_wwz_weights[x]
    else:
        if (arr_p_wwz_s_bkg_bdtR2[x] <= 0.14):
            wwz_yields_r2_bdt[2] += arr_p_wwz_weights[x]
        else:
            wwz_yields_r2_bdt[3] += arr_p_wwz_weights[x]
    #R3 BDT
    if (arr_p_wwz_s_wwz_bdtR3[x] > arr_p_wwz_s_zh_bdtR3[x]):
        if (arr_p_wwz_s_bkg_bdtR3[x] <= 0.13):
            wwz_yields_r3_bdt[0] += arr_p_wwz_weights[x]
        else:
            wwz_yields_r3_bdt[1] += arr_p_wwz_weights[x]
    else:
        if (arr_p_wwz_s_bkg_bdtR3[x] <= 0.14):
            wwz_yields_r3_bdt[2] += arr_p_wwz_weights[x]
        else:
            wwz_yields_r3_bdt[3] += arr_p_wwz_weights[x]

# ZH loop
for x in range (0,arr_p_zh_weights.size):
    #R2 BDT
    if (arr_p_zh_s_wwz_bdtR2[x] > arr_p_zh_s_zh_bdtR2[x]):
        if (arr_p_zh_s_bkg_bdtR2[x] <= 0.13):
            zh_yields_r2_bdt[0] += arr_p_zh_weights[x]
        else:
            zh_yields_r2_bdt[1] += arr_p_zh_weights[x]
    else:
        if (arr_p_zh_s_bkg_bdtR2[x] <= 0.14):
            zh_yields_r2_bdt[2] += arr_p_zh_weights[x]
        else:
            zh_yields_r2_bdt[3] += arr_p_zh_weights[x]
    #R3 BDT
    if (arr_p_zh_s_wwz_bdtR3[x] > arr_p_zh_s_zh_bdtR3[x]):
        if (arr_p_zh_s_bkg_bdtR3[x] <= 0.13):
            zh_yields_r3_bdt[0] += arr_p_zh_weights[x]
        else:
            zh_yields_r3_bdt[1] += arr_p_zh_weights[x]
    else:
        if (arr_p_zh_s_bkg_bdtR3[x] <= 0.14):
            zh_yields_r3_bdt[2] += arr_p_zh_weights[x]
        else:
            zh_yields_r3_bdt[3] += arr_p_zh_weights[x]
            
# BKG loop
for x in range (0,arr_p_bkg_weights.size):
    #R2 BDT
    if (arr_p_bkg_s_wwz_bdtR2[x] > arr_p_bkg_s_zh_bdtR2[x]):
        if (arr_p_bkg_s_bkg_bdtR2[x] <= 0.13):
            bkg_yields_r2_bdt[0] += arr_p_bkg_weights[x]
        else:
            bkg_yields_r2_bdt[1] += arr_p_bkg_weights[x]
    else:
        if (arr_p_bkg_s_bkg_bdtR2[x] <= 0.14):
            bkg_yields_r2_bdt[2] += arr_p_bkg_weights[x]
        else:
            bkg_yields_r2_bdt[3] += arr_p_bkg_weights[x]
    #R3 BDT
    if (arr_p_bkg_s_wwz_bdtR3[x] > arr_p_bkg_s_zh_bdtR3[x]):
        if (arr_p_bkg_s_bkg_bdtR3[x] <= 0.13):
            bkg_yields_r3_bdt[0] += arr_p_bkg_weights[x]
        else:
            bkg_yields_r3_bdt[1] += arr_p_bkg_weights[x]
    else:
        if (arr_p_bkg_s_bkg_bdtR3[x] <= 0.14):
            bkg_yields_r3_bdt[2] += arr_p_bkg_weights[x]
        else:
            bkg_yields_r3_bdt[3] += arr_p_bkg_weights[x]
            
# Now convert the yields lists to numpy arrays
arr_wwz_yields_r2_bdt = np.array(wwz_yields_r2_bdt)
arr_wwz_yields_r3_bdt = np.array(wwz_yields_r3_bdt)
arr_zh_yields_r2_bdt = np.array(zh_yields_r2_bdt)
arr_zh_yields_r3_bdt = np.array(zh_yields_r3_bdt)
arr_bkg_yields_r2_bdt = np.array(bkg_yields_r2_bdt)
arr_bkg_yields_r3_bdt = np.array(bkg_yields_r3_bdt)

# Print out the yields and significances
for y in range(0, arr_wwz_yields_r2_bdt.size):
    print("Signal Region ",y,": ")
    print("---------------------")
    print("R2(R3) WWZ Events = ",arr_wwz_yields_r2_bdt[y]," (",arr_wwz_yields_r3_bdt[y],") ")
    print("R2(R3) ZH Events = ",arr_zh_yields_r2_bdt[y]," (",arr_zh_yields_r3_bdt[y],") ")
    print("R2(R3) BKG Events = ",arr_bkg_yields_r2_bdt[y]," (",arr_bkg_yields_r3_bdt[y],") ")
    print("---------------------")
    
sig_r2_bdt=calculateFOM(arr_wwz_yields_r2_bdt+arr_zh_yields_r2_bdt,arr_bkg_yields_r2_bdt)
sig_r3_bdt=calculateFOM(arr_wwz_yields_r3_bdt+arr_zh_yields_r3_bdt,arr_bkg_yields_r3_bdt)

sig_total_r2 = 0.
sig_total_r3 = 0.

for x in range(0,arr_wwz_yields_r2_bdt.size):
    print("Significance (R2 BDT) in Bin ",x," = ",sig_r2_bdt[x])
    print("Significance (R3 BDT) in Bin ",x," = ",sig_r3_bdt[x])
    sig_total_r2 += (sig_r2_bdt[x])**2.
    sig_total_r3 += (sig_r3_bdt[x])**2

print("Total significance (R2 BDT) = ",np.sqrt(sig_total_r2))
print("Total significance (R3 BDT) = ",np.sqrt(sig_total_r3))
```
